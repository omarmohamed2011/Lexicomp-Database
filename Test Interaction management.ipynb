```python
### Drug-Test Interactions
import json

path_json = 'Notebooks/installations/Data/Results - Final Sprint Assignment/drug-test-conditions.json'
f = open(path_json)

drug_test = json.load(f)
```

```python
drug_test
```

```python
generic_names, lab_names, result_collect = [],[],[]

for drug in drug_test.keys():
    if type(drug_test[drug]) == dict:
        for test in drug_test[drug].keys():
            res = drug_test[drug][test]

            try:
                if res.strip() == 'FALSE':
                    res = 'False results'
                res = res.capitalize(); res = res.strip()
                generic_names.append(drug); lab_names.append(test); result_collect.append(res)
            except:
                res = ''

    else:
        generic_names.append(drug); lab_names.append('No test available'); result_collect.append('No result available')
```

```python
generic_names
```

```python
lab_names
```

```python
df_lab = pd.DataFrame()

df_lab['Generic Name'] = generic_names; df_lab['Test Name'] = lab_names; df_lab['Interaction'] = result_collect
df_lab
```

```python
df_lab.to_csv('Notebooks/installations/Data/Results - Final Sprint Assignment/Drug-Test Interactions.csv',index=False,encoding='utf-8-sig')
```
