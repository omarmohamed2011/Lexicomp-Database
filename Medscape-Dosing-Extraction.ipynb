```python
import re

import pandas as pd
```

```python
dosing_timing_pattern = r"q\d{1,2}-(?:[2-9]|1[0-9]|2[0-4])hr"
text = "The patient should take the medication q8-24hr, and should not exceed q12-24hr. The other medication should be taken q2-8hr, q1-12hr, q8-1|2hr, q90-0hr."

matches = re.findall(dosing_timing_pattern, text)
print(matches)
```

```python
admin_limit_pattern = r"\b\d{1,5}\s+\w+\s+limit\/(?:minute|day)\b"

text = "100 mcg limit/minute, 12 di limit/minute\n5000 kg limit/day"

matches = re.findall(admin_limit_pattern, text, re.MULTILINE)

print(matches)
```

```python
import utils
data_dict=utils.load_json('medscape/Allergy_&_Cold.json')
```

```python
age_ranges=[]
drug_names= []
for class_name,class_data in data_dict.items():

    for drug_name,drug_data in class_data.items():
        drug_names.append(drug_name)
        age_range = drug_data['age_ranges']
        age_ranges.append(age_range)
```

```python
dosingtiming_adult, dosingtiming_pediatric, dosingtiming_geriatric = [],[],[]

for age_range in age_ranges:
    try:
        if 'Adult' in age_range:
            dosingtiming_adult.append(re.findall(dosing_timing_pattern,age_range['Adult']) )
        else:
            dosingtiming_adult.append('None')

    except:
        dosingtiming_adult.append('None')

    try:
        if 'Pediatric' in age_range:
            dosingtiming_pediatric.append(re.findall(dosing_timing_pattern, age_range['Pediatric']) )
        else:
            dosingtiming_pediatric.append('None')

    except:
        dosingtiming_pediatric.append('None')

    try:
        if 'Geriatric' in age_range:
            dosingtiming_geriatric.append(re.findall(dosing_timing_pattern, age_range['Geriatric']) )
        else:
            dosingtiming_geriatric.append('None')

    except:
        dosingtiming_geriatric.append('None')
```

```python
def exact_elems(lst):
    return lst.count(lst[0]) == len(lst)

def first_elem(lst):
    return lst[0]
```

```python
len(dosingtiming_geriatric)
```

```python
for i in range(len(dosingtiming_pediatric)):
    if len(dosingtiming_adult[i]) > 1:
        if exact_elems(dosingtiming_adult[i]):
            dosingtiming_adult[i] = first_elem(dosingtiming_adult[i])

    elif len(dosingtiming_adult[i]) == 1:
        dosingtiming_adult[i] = first_elem(dosingtiming_adult[i])
    else:
        dosingtiming_adult[i] = 'None'

    if len(dosingtiming_pediatric[i]) > 1:
        if exact_elems(dosingtiming_pediatric[i]):
            dosingtiming_pediatric[i] = first_elem(dosingtiming_pediatric[i])

    elif len(dosingtiming_pediatric[i]) == 1:
        dosingtiming_pediatric[i] = first_elem(dosingtiming_pediatric[i])
    else:
        dosingtiming_pediatric[i] = 'None'


    if len(dosingtiming_geriatric[i]) > 1:
        if exact_elems(dosingtiming_geriatric[i]):
            dosingtiming_geriatric[i] = first_elem(dosingtiming_geriatric[i])

    elif len(dosingtiming_geriatric[i]) == 1:
        dosingtiming_geriatric[i] = first_elem(dosingtiming_geriatric[i])
    else:
        dosingtiming_geriatric[i] = 'None'
```

```python
admin_limit_adult, admin_limit_pediatric, admin_limit_geriatric = [],[],[]

for age_range in age_ranges:
    try:
        if 'Adult' in age_range:
            admin_limit_adult.append(re.findall(admin_limit_pattern,age_range['Adult']) )
        else:
            admin_limit_adult.append('None')

    except:
        admin_limit_adult.append('None')

    try:
        if 'Pediatric' in age_range:
            admin_limit_pediatric.append(re.findall(admin_limit_pattern, age_range['Pediatric']) )
        else:
            admin_limit_pediatric.append('None')

    except:
        admin_limit_pediatric.append('None')

    try:
        if 'Geriatric' in age_range:
            admin_limit_geriatric.append(re.findall(admin_limit_pattern, age_range['Geriatric']) )
        else:
            admin_limit_geriatric.append('None')

    except:
        admin_limit_geriatric.append('None')
```

```python
def shuffle_list(elem):
    if elem == [] and len(elem) == 0:
        return 'None'
    else:
        return elem
```

```python
for i in range(633):
    admin_limit_adult[i] = shuffle_list(admin_limit_adult[i])
    admin_limit_pediatric[i] = shuffle_list(admin_limit_pediatric[i])
    admin_limit_geriatric[i] = shuffle_list(admin_limit_geriatric[i])

    dosingtiming_adult[i] = shuffle_list(dosingtiming_adult[i])
    dosingtiming_pediatric[i] = shuffle_list(admin_limit_adult[i])
    dosingtiming_geriatric[i] = shuffle_list(dosingtiming_geriatric[i])
```
