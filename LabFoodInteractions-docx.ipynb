```python
import glob
from src.interactions_utils import *

list_pdfs = glob.glob('Notebooks/installations/Data/Total-Lexi-links-docx/*.docx')[:]
```

```python
list_pdfs
```

```python
from docx import Document

Test_interactions = [];  Food_interactions = [];  titles =[]

for i in range(len(list_pdfs)):
        pdf_path = list_pdfs[i]
        document = Document(pdf_path)
        title = get_title_bypath(pdf_path)

        heading_name1 = 'Test Interactions'
        heading_name2 = 'Food Interactions'

        a = get_heading_text(document.paragraphs, heading_name1)
        b = get_heading_text(document.paragraphs, heading_name2)

        ###############################
        a = a.split(".")
        a = '.'.join(a)

        ###############################
        b = b.split(".")
        b = '.'.join(b)

        titles.append(title); Test_interactions.append(a);  Food_interactions.append(b)
```

```python
import pandas as pd

df_test_inter = pd.DataFrame()
df_test_inter['Drug Name'] = titles
df_test_inter['Test Interactions'] = Test_interactions; df_test_inter['Food Interactions'] = Food_interactions
df_test_inter = df_test_inter[df_test_inter['Drug Name'] != '']

df_test_inter.to_csv('Test Food Interactions.csv')
```
